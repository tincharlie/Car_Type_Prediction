# Replace Missing Data

```python
def replacer(A):
    import pandas as pd
    Q = pd.DataFrame(A.isna().sum(), columns=['MissingVal'])
    R = Q[Q.MissingVal > 0]
    for i in R.index:
        if A[i].dtypes == "object":
            mode = A[i].mode()[0]
            A[i] = A[i].fillna(mode)
        else:
            mean = A[i].mean()
            A[i] = A[i].fillna(mean)
    
```

```python
def anova(A, cat, con):
    from statsmodels.formula.api import ols
    from statsmodels.stats.anova import anova_lm
    rel = con+"~"+cat
    Ols = ols(rel, A)
    model=Ols.fit()
    anova = anova_lm(model)
    ano=pd.DataFrame(anova)
    q = ano['PR(>F)']['Type']
    return round(q,6)
```

# Preprocessing the data

```python
def catcon(A):
    cat = []
    con = []
    for i in A.columns:
        if A[i].dtypes == 'object':
            cat.append(i)
        else:
            con.append(i)
    return cat,con
```

# 1. Logistic Regression

```python
def models(mobj, xtrain, ytrain):
    model = mobj.fit(xtrain,ytrain)
    pr_tr = model.predict(xtrain)
    pr_ts = model.predict(xtest)    
    from sklearn.metrics import accuracy_score
    pd_tr = accuracy_score(ytrain, pr_tr)
    pd_ts = accuracy_score(ytest, pr_ts)
    v = print(mobj,"\nTrain Acc: ", pd_tr, "\nTest Acc: ", pd_ts)
    return v
```

```python
lr = LogisticRegression()
models(lr, xtrain, ytrain)
```

# 2. Decision Tree 

```python
def tune(X, Y, tp, cv, mobj ):
    from sklearn.model_selection import GridSearchCV
    gcv = GridSearchCV(mobj,tp,scoring="accuracy", cv=cv)
    mgcv = gcv.fit(X,Y)
    bestval= mgcv.best_params_
    return bestval
```

```python
dtc = DecisionTreeClassifier(random_state=21)
tp = {"max_depth":range(1,20,1)}
tune(xtrain,ytrain, tp, 5, dtc)
```

```python
dtc = DecisionTreeClassifier(random_state=21, max_depth=6)
models(dtc,xtrain,ytrain)
```

# 3. Random Forest

```python
rfr = RandomForestClassifier(random_state=32)
tp = {"n_estimators":range(1,25,1)}
tune(xtrain, ytrain, tp, 4, rfr)
```

```python
rfr = RandomForestClassifier(random_state=32, n_estimators=8, max_depth=6)
models(rfr, xtrain, ytrain)
```
